## 課題1

各問で示されている図を出力してください。

その際、以下の要件を必ず満たすようにしてください。


1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
size = 5

for i in range(1, size + 1):
    line = ""
    for j in range(i):
        line += "*"
    print(line)
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
size = 5

for i in range(size):
    line = ""
    for j in range(i):
        line += " "
    line += "*"
    print(line)
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5

for i in range(size):
    line = ""
    # 左側のスペース（上の行ほど多くする）
    for j in range(size - 1 - i):
        line += " "
    # 星を1つ
    line += "*"
    print(line)
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5

center = size // 2  # 中央の行

for i in range(size):
    line = ""
    for j in range(size):
        # 左上〜右下、または右上〜左下の対角線上なら "*"
        if j == i or j == size - 1 - i:
            line += "*"
        else:
            line += " "
    print(line)
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 5

for i in range(size):
    line = ""
    # 左側のスペース
    for j in range(size - 1 - i):
        line += " "
    # 星の数（1,3,5,... と奇数で増える）
    for j in range(2 * i + 1):
        line += "*"
    print(line)
````
